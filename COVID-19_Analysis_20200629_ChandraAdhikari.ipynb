## Number of Days vs reference  numbers of cases for each countries

```python
def get_indices(N):
    index_N={}
    for i in range(1, N_countries):
        cases=0
        for j in range(len(data_death['China'])):
            cases += data_case[countries[i]][j]
            if cases >= N:
                index_N[countries[i]] = j
                break
    return index_N

Ncases = [1000, 10000, 50000, 100000, 500000, 1000000]

for i, N in enumerate(Ncases):
    idx_N=get_indices(N)
    print ("Days to reach", N, 'case:', idx_N)
```

```python
def Plot_Milestones():
    Nmilestones = len(Ncases)
    Milestones=np.zeros((Nmilestones, N_countries-1), dtype=int)

    for i, Nm in enumerate(Ncases):
        idx_N=get_indices(Nm)
        Milestones[i,:] = list(idx_N.values())
        
    for i in range(N_countries-1):
        xx=range(Nmilestones)
        plt.plot(xx, Milestones[:,i], '-d', label=countries[i+1])
        plt.legend(loc='center left', frameon=False, fontsize=11,bbox_to_anchor=(1, 0.5))
        plt.xticks(xx, Ncases)
        plt.xlabel("Total Number of Cases", fontsize=13)
        plt.ylabel("Number of Days", fontsize=13)
Plot_Milestones()
```


# Deaths per day
 

```python
#Gaussian Fit
from scipy.optimize import curve_fit
def Gauss(x, mu, sigma, C):
    return C*1./(sigma * np.sqrt(2 * np.pi)) * np.exp( - (x - mu)**2 / (2 * sigma**2) )


data_c_day = data_case.diff().fillna(data_case)
data_d_day = data_death.diff().fillna(data_case)

def PlotPerDay():
    plt.figure(1, figsize=(18, 6))
    for i in range(1, N_countries-4):    
        plt.subplot(240+i)
        plt.title(countries[i], fontsize=14, color='k')
        xx = np.array(range(len(data_c_day[countries[i]].index)) )
        yy = data_c_day[countries[i]]
        plt.plot(xx, yy, 'd')
        plt.xlabel("Number of Days", fontsize=12)
        plt.ylabel("Cases per Day", fontsize=12)
        
        plt.subplots_adjust(bottom=.0001, top=.9999, hspace=0.5, wspace=0.325)
        
        mean = sum(xx * yy) / sum(yy)
        sigma = np.sqrt(sum(yy * (xx - mean)**2) / sum(yy))

        #print (mean, sigma)
        popt, pcov =curve_fit(Gauss, xx ,yy, p0=[mean, sigma, max(yy)])
        #print (popt)
        plt.plot(xx, Gauss(xx, *popt), 'r-')

PlotPerDay()
```
